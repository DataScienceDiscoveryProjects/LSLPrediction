```python
import pandas as pd
import random

denver = pd.read_csv('/work/denver.csv') 
dc = pd.read_csv('/work/dc.csv')
cinc = pd.read_csv('/work/cinc.csv')
trent = pd.read_csv('/work/trent.csv')
pitt = pd.read_csv('/work/pitt.csv')
```

```python
#Total initial rows
total_rows = len(denver) + len(dc)+ len(cinc) + len(trent) + len(pitt)
print(total_rows)

#number random rows to drop
print(total_rows - 300000) 
#need 300000/5 = 60000 rows per dataset
```

```python
#DENVER
denver = denver[['prem_full_addr','lsli_final_lead_cat' ]]
denver_lead = pd.get_dummies(denver['lsli_final_lead_cat'])

denver = pd.concat([denver, denver_lead], axis = 1).drop('lsli_final_lead_cat', axis = 1)
denver['City'] = 'Denver'
denver['State'] = 'Colorado'
denver = denver.rename(columns = {'prem_full_addr': 'Address', 'Confirmed lead service line; included in replacement program.': 'Lead', 'No lead service line.': 'No Lead'})
denver['Address'] = denver['Address'].drop_duplicates()
denver = denver.dropna()
num_rows_to_drop = len(denver) - 60000
# Generate a list of random row indices to drop
random_indices = random.sample(denver.index.tolist(), num_rows_to_drop)

# Drop the randomly selected rows
denver = denver.drop(random_indices)

lead_count = len(denver[denver['Lead']==1])

print('Total row count:')
print(len(denver))
print('Lead count:')
print(lead_count)
print('dupes:')
print(denver[denver['Address'].duplicated()])
denver 


#0 na values for lsli_final_lead_cat
#important columns: prem_full_addr, parcelyearbuilt, servicearea, mediumhouseholdincome2015, point_x,
# point_y, denver_district, lsli_final_lead_cat, NHNAME, Length
```

```python
#DC
dc.head()

dc = dc.dropna(subset=['PublicSymbol'])

num_rows_to_drop = len(dc) - 60000


dc = dc[['PremiseAddress','PublicSymbol']]
dc_lead = pd.get_dummies(dc['PublicSymbol'])
dc = pd.concat([dc, dc_lead], axis = 1).drop('PublicSymbol', axis = 1)
dc['City'] = 'DC'
dc['State'] = 'Washington'
dc= dc.rename(columns = {'PremiseAddress': 'Address', 'NonLead': 'No Lead'}).drop('Unknown', axis = 1)
dc['Address'] = dc['Address'].drop_duplicates()
dc = dc.dropna()
num_rows_to_drop = len(dc) - 60000
# Generate a list of random row indices to drop
random_indices = random.sample(dc.index.tolist(), num_rows_to_drop)

# Drop the randomly selected rows
dc = dc.drop(random_indices)
lead_count = len(dc[dc['Lead']==1])
print('Total row count:')
print(len(dc))
print('Lead count:')
print(lead_count)
print('dupes:')
print(dc[dc['Address'].duplicated()])
dc 


#important columns: PremiseAddress, PublicServiceMaterialType, PublicSymbol, PrivateServiceMaterialType, PrivateSymbol
#
```

```python
#CINCINATTI

cinc = cinc.dropna(subset=['CITYSERVICEMAT'])

cinc = cinc[['PROPADDRESS','CITYSERVICEMAT']]
cinc_lead = pd.get_dummies(cinc['CITYSERVICEMAT'])
cinc = pd.concat([cinc, cinc_lead], axis = 1).drop('CITYSERVICEMAT', axis = 1)
cinc['City'] = 'Cincinnati'
cinc['State'] = 'Ohio'
cinc = cinc.rename(columns = {'PROPADDRESS': 'Address', 'LEAD': 'Lead'})
cinc['Address'] = cinc['Address'].drop_duplicates()
cinc = cinc.dropna()
num_rows_to_drop = len(cinc) - 60000
# Generate a list of random row indices to drop
random_indices = random.sample(cinc.index.tolist(), num_rows_to_drop)

# Drop the randomly selected rows
cinc = cinc.drop(random_indices)

# Define a mapping dictionary or function
mapping = {1: 0, 0: 1}

# Use the map function to create a new column
cinc['No Lead'] = cinc['Lead'].map(mapping)
cinc = cinc[['Address', 'Lead', 'No Lead', 'City', 'State']]
lead_count = len(cinc[cinc['Lead']==1])
print('Total row count:')
print(len(cinc))
print('Lead count:')
print(lead_count)
print('dupes:')
print(cinc[cinc['Address'].duplicated()])
cinc
#PROPADDRESS, TAPLOCATION, CITYSERVICESIZE, CITYSERVICEMAT, PVTSERVICESIZE, PVTSERVICEMAT
```

```python
#TRENTON

trent = trent.dropna(subset=['TWWFinalMat'])

trent = trent[['ServiceLocation','TWWFinalMat']]
trent_lead = pd.get_dummies(trent['TWWFinalMat'])
trent = pd.concat([trent, trent_lead], axis = 1).drop('TWWFinalMat', axis = 1)
trent['City'] = 'Trenton'
trent['State'] = 'New Jersey'
trent= trent.rename(columns = {'ServiceLocation': 'Address', 'LEAD': 'Lead', 'Not Lead': 'No Lead'})
trent = trent[['Address', 'Lead', 'No Lead', 'City', 'State']]

trent['Address'] = trent['Address'].drop_duplicates()
trent = trent.dropna()

num_rows_to_drop = len(trent) - 60000
# Generate a list of random row indices to drop
random_indices = random.sample(trent.index.tolist(), num_rows_to_drop)

# Drop the randomly selected rows
trent = trent.drop(random_indices)

lead_count = len(trent[trent['Lead']==1])
print('Total row count:')
print(len(trent))
print('Lead count:')
print(lead_count)
print('dupes:')
print(trent[trent['Address'].duplicated()])

trent

# important columns: ServiceLocation, TWWFinalMat
```

```python
#PITTSBURGH

pitt = pitt.dropna(subset=['FinalReportedMaterialPublic'])

pitt['Address'] = pitt[pitt['Address'] !='EWING STREET']
pitt['Address'] = pitt['Address'].astype(str) + ' ' + pitt['Street']
pitt['City'] = 'Pittsburgh'
pitt['State'] = 'Pennsylvania'


pitt_lead = pd.get_dummies(pitt['FinalReportedMaterialPublic'])
pitt = pd.concat([pitt, pitt_lead], axis = 1).drop('FinalReportedMaterialPublic', axis = 1)

pitt = pitt[['Address', 'Lead', 'NonLead', 'City', 'State']]
pitt = pitt.rename(columns = {'NonLead': 'No Lead'})
pitt = pitt.dropna(subset=['Address'])

pitt['Address'] = pitt['Address'].drop_duplicates()
pitt = pitt.dropna()
num_rows_to_drop = len(pitt) - 60000
# Generate a list of random row indices to drop
random_indices = random.sample(pitt.index.tolist(), num_rows_to_drop)

# Drop the randomly selected rows
pitt = pitt.drop(random_indices)

lead_count = len(pitt[pitt['Lead']==1])
print('Total row count:')
print(len(pitt))
print('Lead count:')
print(lead_count)
print('dupes:')
print(pitt[pitt['Address'].duplicated()])

pitt
```

```python
#Fuse all data together
cross_regional = pd.concat([denver, dc, cinc, trent, pitt], ignore_index=True)
cross_regional[cross_regional['Lead']==1]
#cross_regional.to_csv('CrossRegionalData.csv', sep=',', index=False, encoding='utf-8')
```
