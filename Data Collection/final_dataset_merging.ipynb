```python
import pandas as pd
```

```python
data = pd.read_csv('/Users/andrewkerr/Downloads/CrossRegionalData (2).csv')
data
```

```python
pitt = data[data['City'] == 'Pittsburgh']
pitt
```

```python
pitt[pitt['Address'].str.len() < 20]['Address']
```

```python
def remove_duplicates(row):
    words = row.split()
    unique_words = []
    for word in words:
        if word not in unique_words:
            unique_words.append(word)
    return ' '.join(unique_words)
```

```python
remove_duplicates('900 5TH AVE 5TH AVE')
```

```python
pitt['Address'] = pitt['Address'].apply(lambda x: remove_duplicates(x))
pitt
```

```python
without_pitt = data[data['City'] != 'Pittsburgh']
without_pitt
```

```python
result = pd.concat([without_pitt, pitt])
result
```

```python
census = pd.read_csv('/Users/andrewkerr/Downloads/census_normalized_with_year.csv')
census
```

```python
census['abbreviatedAddress'] = census['abbreviatedAddress'].str.lower()
census
```

```python
result['Address'] = result['Address'].str.lower()
result['Address'] 
```

```python
merge = result.merge(census, left_on = 'Address', right_on = 'abbreviatedAddress')
merge = merge.drop(columns = ['No Lead', 'Unnamed: 0', 'abbreviatedAddress', 'City', 'State', 'streetAddress'])
merge
```

```python
merge.to_csv('final_dataset.csv', index=False)
```
